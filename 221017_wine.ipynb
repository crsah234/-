```python
# 열 1ㅡ2ㅡ3 을 학습시켜서 열4를 맞추기.
input_data = wine[['alcohol', 'sugar', 'pH']]
input_data # <class 'pandas.core.frame.DataFrame'>
    

# numpy로 변환
data=input_data.to_numpy()
data   # <class 'numpy.ndarray'>
```

```python
# 레이블 셋 편집
target = wine['class'].to_numpy() # 위에서 리스트로 가져왓엇으나 지금은한개만 가져오면 되기에 값만 넣기
target # <class 'numpy.ndarray'>
print(target.shape) # (6497,)

# data -> feature값, target -> 레이블 값
```

```python
# 학습 데이터셋과 시험 데이터 셋으로 나누기.
```

```python
from sklearn.model_selection import train_test_split

# 각각 data, target을 나눠주기, test_size는 default가 3:!(학습:시험) 임. 
train_input, test_input,train_target, test_target = train_test_split(data, target, test_size=0.2,random_state=42)

print(train_input.shape) # (5197, 3)
print(test_input.shape) # (1300, 3)
```

```python
from sklearn.preprocessing import StandardScaler # 클래스 import 

ss = StandardScaler() # ss 객체 생성
# 데이터 피팅, 변환 2가지 절차가 필요함
ss.fit(train_input) # fit()을 통해 학습데이터셋의 평균, 표준편차 계산

# 표준점수로 변환한 두 변수
train_scaled = ss.transform(train_input) # z-score로 변형시키고, 변형결과를 저장
test_scaled=ss.transform(test_input) # 위와 동일
```

```python
from sklearn.neighbors import KNeighborsClassifier # 클래스 호출

kn=KNeighborsClassifier()
kn.fit(train_scaled,train_target)
print("학습 데이터 셋 : ",kn.score(train_scaled,train_target)) # 모의고사
print("시험 데이터 셋 : ",kn.score(test_scaled,test_target)) # 수능

```

```python
print(kn.n_neighbors) # KNN은 가장 가까운 데이터를 참고하여 분류.
```

```python
# 최적의 K값을 찾아내는 방법 2~100
score=0

# k값의 따른 변화 추이도 그래프로 그려보기
hist=[]

for k in range(2,101):
    kn.n_neighbors=k
    
    tmp= kn.score(test_scaled, test_target)
    hist.append(round(tmp*100,2))
    if tmp>score:
        score=tmp
        best_k=k
        
print("최적의 k : ",best_k)
print("최고 정확도 : ",round(score*100,2),'%')

import matplotlib.pyplot as plt
plt.figure()
plt.plot(range(2,101),hist)
plt.xlabel('k')
plt.ylabel('Accuracy(%)')
plt.show()
```
